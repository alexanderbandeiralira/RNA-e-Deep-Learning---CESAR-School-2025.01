# Regressão Linear com PyTorch

## Objetivos

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se:
- a função de perda MSE do PyTorch,
- treinamento dos parâmetros via gradiente descendente usando o otimizador.
- A rede é criada com uma camada nn.Linear()

## Importação dos pacotes

````python
%matplotlib inline
import torch
from torch import nn, optim
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

torch.manual_seed(1234)
````

## Leitura dos dados

````python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
````

### Normalização dos dados

````python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
````

````python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
````

````python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
````

## Criação do modelo da rede

````python
model = torch.nn.Linear(2, 1, bias=False)
````

### Verificando a inicialização dos parâmetros

````python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight.data
````

### Testando o predict da rede

````python
model(Variable(torch.ones((5,2))))
````

````python
model(x_train_bias)
````

## Treinamento

### Definindo função de perda e otimizador

````python
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.3)
````

### Laço de treinamento

````python
num_epochs = 120
w0_list = []
w1_list = []
for epoch in range(num_epochs):
    inputs = x_train_bias
    target = y_train

    # forward - predict
    out = model(inputs)

    w0_list.append(model.weight.data[0][0].item())
    w1_list.append(model.weight.data[0][1].item())

    #loss cálculo da função de perda
    # loss = torch.mean((y_pred - target)**2)
    loss = criterion(out, target)

    # backward e otimizador
    optimizer.zero_grad()
    loss.backward()
    # model.weight.data = model.weight.data - learning_rate * model.weight.grad.data
    optimizer.step()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
````

## Avaliação

````python
y_pred = model(x_train_bias)
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
````

- 2) Faça igual o exercício do notebook anterior, de plotar um gráfico scatterplot para
   mostrar a evolução dos parâmetros durante o treinamento pelo gradiente descendente.

````python
# todo
x_bias_t = torch.t(x_train_bias)
w_opt = (torch.inverse(x_bias_t.mm(x_train_bias)).mm(x_bias_t)).mm(y_train)


w0_opt = w_opt[0].item()  # w0 (Bias)
w1_opt = w_opt[1].item()  # w1 (Slope)

plt.figure(figsize=(8, 6))

# Trajetória (Pontos)
plt.scatter(w0_list, w1_list, c='blue', marker='.', label='Trajetória do Gradiente')

# Trajetória (Setas)
w0_old = None
w1_old = None
for (w0, w1) in zip(w0_list, w1_list):
    if w0_old is not None:
        plt.arrow(w0_old, w1_old, w0 - w0_old, w1 - w1_old,
                  head_length=0.002, head_width=0.002, shape='full',
                  length_includes_head=True, color='gray', alpha=0.5)
    w0_old, w1_old = w0, w1

# Início e Fim do Treinamento
plt.scatter(w0_list[0], w1_list[0], c='green', marker='o', s=100, label='Início do Treinamento')
plt.scatter(w0_list[-1], w1_list[-1], c='red', marker='o', s=100, label='Fim do Treinamento')

# Solução Analítica (Ótimo)
plt.scatter(w0_opt, w1_opt, c='black', marker='*', s=300, label='Solução Analítica (Ótimo)')

plt.title('Convergência dos Parâmetros do Modelo (SGD Optimizer)')
plt.xlabel('$w_0$ (Bias)')
plt.ylabel('$w_1$ (Slope)')
plt.legend()
plt.grid(True)
plt.savefig('parameter_evolution_sgd.png')
plt.show()
````
